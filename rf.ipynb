```python
from sklearn import preprocessing 
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import RandomizedSearchCV
```

```python
df = pd.read_csv('Training.csv',on_bad_lines='skip',low_memory=False)
```

```python
df.dropna()
```

```python
df.isnull().values.any()
```

```python
df.head()
```

```python
df['prognosis'].unique()
```

```python
encoder = LabelEncoder()
df["prognosis"] = encoder.fit_transform(df["prognosis"])
df = df.sample(frac=1).reset_index(drop=True)
#Y
Y = df['prognosis']
Y = np.asarray(Y)
#X 
X =df.drop(['prognosis'], axis=1)
X = np.asarray(X)

print(X.shape)
print(Y.shape)
```

```python
x_train, x_test,y_train, y_test = train_test_split(X,Y ,random_state=42, test_size=0.2, shuffle=True)
```

```python
n_estimators = np.linspace(100, 1000, 100, dtype=int)
max_depth = [1, 5, 10, 20, 50, 75, 100, 150, 200]
min_samples_split = [1, 2, 5, 10, 15, 20, 30]
min_samples_leaf = [1, 2, 3, 4]
bootstrap = [True, False]

criterion=['gini', 'entropy']
random_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap,
               'criterion': criterion}
```

```python
rf_base = RandomForestClassifier()
rf_random = RandomizedSearchCV(estimator = rf_base,param_distributions = random_grid,verbose=2,random_state=42)
rf_random.fit(x_train, y_train)
```

```python
rf_random.best_params_
```

```python
clf = RandomForestClassifier(n_estimators=963,min_samples_split= 10,min_samples_leaf=3,max_depth= 75,criterion='gini',bootstrap= True )
clf.fit(x_train,y_train)
pred=clf.predict(x_test) 
```

```python
result = confusion_matrix(y_test, pred)
```

```python
print('Matrix: ')
print(result)
```

```python
result1 = classification_report(y_test, pred)
print('Report:')
print (result1)
```

```python
result2 = accuracy_score(y_test,pred)
print("Accuracy",result2)
```
